### U-Matrix e Mapa de Hits (Grade 10x10)
A U-Matrix ajuda a visualizar as fronteiras de distância na grade, enquanto o mapa de Hits mostra a densidade de amostras por neurônio.

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 7), dpi=150)
axes[0].imshow(Image.open("../outputs/figures/umatrix_10x10.jpg"))
axes[0].axis('off')
axes[0].set_title("U-Matrix 10x10", fontsize=14, fontweight='bold')

axes[1].imshow(Image.open("../outputs/figures/umatrix_hits_10x10_with_hits.jpg"))
axes[1].axis('off')
axes[1].set_title("Mapa de Hits (Amostras)", fontsize=14, fontweight='bold')

plt.show()
```
