## 7. Discussões e Conclusões
1. **Mapeamento de séries temporais**: Os mapas SOM de $7\times7$ e $10\times10$ alcançaram o maior alinhamento de clusters com os rótulos reais ($NMI \approx 0.81$, $ARI \approx 0.66$), superando os métodos clássicos direct K-Means ($ARI \approx 0.56$).
2. **Padrão de classes**: Conforme a literatura, classes com tendências similares (ex: Increasing Trend e Upward Shift) ficaram próximas ou compartilharam vizinhança nas U-Matrices.
3. **Extensão Textual**: O uso de embeddings profundos contextuais (Sentence-BERT) permitiu uma organização topológica semântica significativamente superior ($ARI \approx 0.27$, $NMI \approx 0.34$) se comparado à contagem de termos TF-IDF ($ARI \approx 0.19$, $NMI \approx 0.26$).
4. **Aplicações Futuras**: Demonstramos que a IntraSOM pode ser estendida com sucesso para triagem de textos, ementas de processos jurídicos e organização automática de memórias técnicas.
